### Quality

#### Remove retweeted tweets from `df`

We want only original dog ratings (no retweets) so we need to remove all rows that have values in retweeted_status_id, retweetd_status_user_id, retweeted_status_timestamp columns.

#### `df`: Combine Key Word columns into one "catch-all" column

##### Define

Doggo/Floofer/Pupper/Puppo Columns should be concatenated into one dog_stage column. Start by combining all doggo, floofer, pupper, poppu columns into one dog_stage column, then clean that column for only the key word we're looking for. From there, drop original four columns.

##### Code

```python
# create concatenated names column
df_clean['good_dog_stage'] = df_clean.doggo + df_clean.floofer + df_clean.pupper + df_clean.puppo
df_clean.loc[df_clean.good_dog_stage == 'doggopupper', 'good_dog_stage'] = 'doggo,pupper'
df_clean.loc[df_clean.good_dog_stage == 'doggopuppo', 'good_dog_stage'] = 'doggo,puppo'
df_clean.loc[df_clean.good_dog_stage == 'doggofloofer', 'good_dog_stage'] = 'doggo,floofer'
```

```python
# replace the concatenated names with correct names
df_clean.good_dog_stage.replace('NoneNoneNoneNone', 'unavailable',inplace = True)
df_clean.good_dog_stage.replace('NoneNonepupperNone', 'pupper', inplace = True)
df_clean.good_dog_stage.replace('doggoNoneNoneNone', 'doggo', inplace = True)
df_clean.good_dog_stage.replace('NoneNoneNonepuppo', 'puppo', inplace = True)
df_clean.good_dog_stage.replace('doggoNonepupperNone', 'doggo, pupper', inplace = True)
df_clean.good_dog_stage.replace('NoneflooferNoneNone', 'floofer', inplace = True)
df_clean.good_dog_stage.replace('doggoflooferNoneNone', 'doggo, floofer', inplace = True)
df_clean.good_dog_stage.replace('doggoNoneNonepuppo', 'doggo, pupper', inplace = True)
```

```python
# drop unnecessary columns
df_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis = 1, inplace = True)
```

#### Insight #3: Average Rating

```python
# original mean before cleaning
df.rating_numerator.mean()
```

```python
# mean after cleaning
twitter_archive_master.rating_numerator.mean()
```

```python
# median would be used if we still believe there are outliers
twitter_archive_master.rating_numerator.median()
```

- There are two ways to interpret the "average" rating of these doggos. The first inclination is to calculate for the mean, which outputs an average rating of 12.25. However, considering that we have outliers in this dataset - specifically, the maximum of 1776 that is clearly going to skew our data - the best measure of spread here would be the median. The best way to represent the average rating of these dogs is in the median, which is an 11/10. After cleaning the data, we see that our new mean is 10.6, which is now much closer to our median.
